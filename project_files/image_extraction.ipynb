### Processing the FoodSeg103 dataset

```python
food_mask_path = "./data/FoodSeg103/Images/ann_dir"
food_image_path = "./data/FoodSeg103/Images/img_dir"
food_processed_path = "./data/FoodSeg103/Processed"
```

```python
categories = {}
for category in np.genfromtxt("./data/FoodSeg103/category_id.txt", dtype='str'):
    categories[int(category[0])] = {
        "label": category[1].lower(),
        "encounters": 0
    }
```

```python
splits = ["train"]

index = 0

for split in splits:
    for file in os.listdir(f"{food_mask_path}/{split}"):
        if file.endswith(".png"):
            masks = skimage.io.imread(f"{food_mask_path}/{split}/{file}")
            
            image_id = file.split(".")[0]
            image = skimage.io.imread(f"{food_image_path}/{split}/{image_id}.jpg")
            
            for grey_value in np.unique(masks):  
                if grey_value == 0:
                    continue

                mask = deepcopy(masks)
                mask[mask != grey_value] = 0
                mask[mask == grey_value] = 1
                image = cv2.cvtColor(image, cv2.COLOR_RGB2RGBA)

                result = np.zeros_like(image, dtype=np.uint8)

                # Update every channel
                for i in range(image.shape[-1]):
                    result[..., i] = np.multiply(image[..., i], mask)

                # Find the bounding box
                bbox = np.argwhere(mask)
                (y_start, x_start), (y_stop, x_stop) = bbox.min(0), bbox.max(0) + 1 
                result = result[y_start:y_stop, x_start:x_stop]

                # Pad image
                result = np.pad(result, ((10, 10), (10, 10), (0, 0)))

                filename = f"{categories[grey_value]['label']}{categories[grey_value]['encounters']}.png"
                categories[grey_value]['encounters'] += 1
                
                dirs = f"{food_processed_path}/{split}/{categories[grey_value]['label']}"
                if not os.path.exists(dirs):
                    os.makedirs(dirs)
                    
                # Check if the model accepts transparency -- for InceptionV3 it is 3

                transparency = result[..., 3] == 0
                result[transparency] = [255, 255, 255, 255]
                result = cv2.cvtColor(result, cv2.COLOR_BGRA2BGR)
                
                result = np.array(result)
                # Data augmentation?
            
                result = skimage.transform.resize(result, (300, 300), anti_aliasing=True)
                skimage.io.imsave(f"{dirs}/{filename}", result)
```
